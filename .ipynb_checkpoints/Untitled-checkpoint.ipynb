```python
student.columns = ['school', 'sex', 'age', 'address', 'familysize', 'Parentstatus', 'Motherseducation', 'Fatherseducation',
       'Mjob', 'Fjob', 'reason', 'guardian', 'traveltime', 'studytime',
       'failures', 'schoolsup', 'famsup', 'paid', 'activities', 'nursery',
       'higher', 'internet', 'romantic', 'famrel', 'freetime', 'goout', 'Dalc',
       'Walc', 'health', 'absences', 'Gradeone', 'Gradetwo', 'finalscore']
```

```python
student.columns
```

```python
# converting finalscore into categorical variable - # Excellent - 15 to 20 Good - 10 to 14 Fair - 0 to 9
student['finalgrade'] = "na"
student.loc[(student.finalscore >= 15) & (student.finalscore <= 20), 'finalgrade'] = 'excellent' 
student.loc[(student.finalscore >= 10) & (student.finalscore <= 14), 'finalgrade'] = 'good' 
student.loc[(student.finalscore >= 0) & (student.finalscore <= 9), 'finalgrade'] = 'fair' 
student.head(5)
```

```python
# look for missing values
student.isnull().any()
```

```python
import seaborn as sns
sns.set_style('whitegrid')
```

```python
# Final Grade Countplot
plt.figure(figsize=(8,6))
sns.countplot(student.finalgrade, order=["fair","good","excellent"], palette='Set1')
plt.title('finalgrade - Number of Students',fontsize=20)
plt.xlabel('finalgrade', fontsize=16)
plt.ylabel('Number of Student', fontsize=16)
plt.savefig('finalgrade.png')
#sns.plt.show()
```
